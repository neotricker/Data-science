```python
import pandas
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [1,2,3,4]
y = [2,6,12,20]
plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [2,6,12,20]
plt.plot(x,y, label="myPlot", color="red", linewidth=5)
plt.legend()
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
plt.plot([0.3, 0.8, 2.2, 4.0], [11, 25, 9, 26], color='darkgreen', marker='o')
plt.xlim(0.5, 4.5)
plt.show()

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(2,2,1)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.set_xlim(0.5, 4.5)

ax = fig.add_subplot(2,2,2)
ax.plot([1, 2, 2, 4], [10, 20, 25, 30], color='red', linewidth=3)
ax.set_xlim(0.5, 4.5)

ax = fig.add_subplot(2,2,3)
ax.plot([2, 2, 3, 4], [10, 20, 25, 30], color='green', linewidth=3)
ax.set_xlim(0.5, 4.5)

ax = fig.add_subplot(2,2,4)
ax.plot([1, 1, 3, 4], [10, 20, 25, 30], color='black', linewidth=3)
ax.set_xlim(0.5, 4.5)
ax.set(title="A title", xlabel="axis1", ylabel="axis2")
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3, label="a", linestyle='--')
ax.plot([0.3, 0.8, 2.2, 4.0], [11, 25, 9, 26], color='darkgreen', marker='+', label="b")
ax.set_xlim(0.5, 6)
plt.legend()
plt.savefig("wykres.png")
```

```python
fig = plt.figure(figsize=(16,8))
ax = fig.add_subplot(2,4,1)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.set_title("line plot")

ax = fig.add_subplot(2,4,2)
ax.scatter([1, 2, 3, 4], [10, 20, 25, 30], color='red', marker='<')
ax.scatter([1, 2, 3, 4], [30, 25, 15, 10], color='brown', marker='x')
ax.set_title("scatter plot")

ax = fig.add_subplot(2,4,3)
ax.bar(['a','b','c'],[3,4,5], color='green')
ax.bar(['d','e'],[3,4], color='blue')
ax.axhline(2, color='red')
ax.set_title("bar chart")

ax = fig.add_subplot(2,4,4)
ax.barh(['a','b','c'],[3,4,5], color='green')
ax.barh(['d','e'],[3,4], color='blue')
ax.set_title("vertical bar chart")

ax = fig.add_subplot(2,4,5)
ax.fill([1,1,2,2], [1,2,2,1], "red", [1.5, 2.2, 2.6], [1.2, 1.8, 1.2], "green")
ax.set_title("fill")

ax = fig.add_subplot(2,4,6)
ax.stackplot([1,2,3],[3,4,5],[5,6,7], [8,9,10])
ax.set_title("stackplot")

ax = fig.add_subplot(2,4,7)
data = [1,2,3,2,1,2,3,4,2,2,3,3,2,1,2,3,4,2,2,3,3,5]
ax.hist(data)
ax.set_title("histogram")

ax = fig.add_subplot(2,4,8)
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
ax.set_title("pie chart")

plt.show()
```
